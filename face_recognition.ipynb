```python
# Training Model

import tensorflow as tf
from keras import layers, models

# Define the CNN model
model = models.Sequential()

# Convolutional layers
model.add(layers.Conv2D(32, (1, 1), activation='relu', input_shape=(62, 47, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Flatten layer to transition from convolutional to fully connected layers
model.add(layers.Flatten())

# Dense (fully connected) layers
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.5))  # Dropout for regularization
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(5, activation='softmax'))  # Assuming num_classes is the number of people/classes for face recognition

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Display the summary of the model architecture
model.summary()
```

```python
# Fit traing set into model.

model.fit(X_images, y_train, epochs=100)
```

```python
# Predict with the test data.

predicted=model.predict(X_test)
```
